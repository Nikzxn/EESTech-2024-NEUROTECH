```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from catboost import CatBoostClassifier
```

```python
data=pd.read_csv("future_labels.csv")
```

```python
data.head()
```

```python
X,y=data.drop(["status","future_status","Дата и время","date","hour","tractor"],axis=1),data["future_status"]
```

```python
data["future_status"].value_counts()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,stratify=y,random_state=42,test_size=0.2)
```

```python
from imblearn.over_sampling import RandomOverSampler
ros=RandomOverSampler(random_state=42)
X_train_ros,y_train_ros=ros.fit_resample(X_train,y_train)
```

```python
cbc=CatBoostClassifier(task_type="GPU")
```

```python
cbc.fit(X_train,y_train,verbose=50)
```

```python
pred=cbc.predict(X_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,pred))
```
